```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
Items = pd.read_csv('Items.csv')
customers = pd.read_csv('customers.csv')
order_item = pd.read_csv('order_item.csv')
orders = pd.read_csv('orders.csv')
```

```python
dup_counts = Items.groupby('item_name')['id'].count()
dup_counts[dup_counts > 1]
```

```python
dup = Items[Items['item_name'].duplicated(keep=False)].copy()
dup
```

```python
dup['non_nan_count'] = dup.notna().sum(axis=1)
dup.sort_values(['item_name', 'non_nan_count'], ascending=[True, False], inplace=True)
dup
```

```python
kept = dup.drop_duplicates(subset=['item_name'], keep='first')
removed = dup[~dup['id'].isin(kept['id'])] 
removed
```

```python
items_clean = Items[~Items['id'].isin(removed['id'])]
items_clean
```

```python
dup_counts = items_clean.groupby('item_name')['id'].count()
dup_counts[dup_counts > 1]
```

```python
items_clean['item_name'].duplicated().sum()
```

```python
removed_items = Items[~Items['id'].isin(items_clean['id'])]
merged = removed_items.merge(items_clean, on='item_name', suffixes=('_removed', '_kept'))
item_map = dict(zip(merged['id_removed'], merged['id_kept']))
order_item['item_id'] = order_item['item_id'].replace(item_map)
```

```python
order_counts = orders['customer_id'].value_counts()

customers_to_keep = order_counts[order_counts >= 3].index

customers_clean = customers[customers['id'].isin(customers_to_keep)]

orders_clean = orders[orders['customer_id'].isin(customers_clean['id'])]

order_item_clean = order_item[order_item['order_id'].isin(orders_clean['order_id'])]
order_item_clean
```
